```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(20,10))
cor = df.corr()
sns.set(font_scale=1.4)
sns.heatmap(cor, annot=True, cmap=plt.cm.Blues)
plt.show()
```

```python
sns.set(font_scale=4) 
sns.pairplot(data=df, hue="Product", height=10, aspect=0.6)
```

```python
df.dtypes
```

```python
Gender = pd.crosstab(df["Product"],df["Gender"])
Gender
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15,10))

#sns.barplot(ax=axes[0, 0], x='Gender', y='Income',data=df, hue = 'Product',ci=None)

plt.subplots_adjust(hspace = 0.5)
axe = axes[1][0]
sns.countplot(x='Product' , hue='Gender', data=df,ax=axes[1][0]).set_title("Gender by Product") # Impact of gender on product-visualisation

sns.barplot(x='Gender', y='Income',data=df, hue = 'Product',ci=None, ax=axes[1][1]).set_title("Gender by Income")

sns.barplot(x='MaritalStatus', y='Income',data=df, hue = 'Product',ci=None, ax=axes[0][1]).set_title("MaritalStatus by Product")

sns.barplot(x='Fitness', y='Income',data=df, hue = 'Product',ci=None, ax=axes[0][0]).set_title("MaritalStatus by Product")
```
